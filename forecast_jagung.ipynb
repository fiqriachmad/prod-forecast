```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv(r'direktori file.csv')
```

```python
data
```

```python
data.info()
```

```python
data_trans = data.T
```

```python
data_trans
```

```python
new_header = data_trans.iloc[0] #grab the first row for the header
data_trans = data_trans[1:] #take the data less the header row
data_trans.columns = new_header #set the header row as the df header
```

```python
data_trans.index.name
data_trans.index.name = 'Tahun'
data_trans.index.name
```

```python
data_trans.replace('-', '0', inplace=True)
data_trans = pd.DataFrame(data=data_trans)
data_trans = data_trans.astype(int)
```

```python
data_trans
```

```python
data_trans.info()
```

```python
data_aceh = pd.DataFrame(data=data_trans['ACEH'])
data_aceh = data_aceh.astype(int)
data_aceh.dtypes
data_aceh.info()
```

```python
data_aceh.plot(figsize=(15,12))
```

```python
data_trans.plot(figsize=(15,12))
```

```python
data_aceh
```

```python
from statsmodels.tsa.arima_model import ARIMA
from sklearn.metrics import mean_squared_error
```

```python
model_aceh = ARIMA(data_aceh, order=(1,1,1))
model_arima_aceh = model_aceh.fit()
```

```python
model_arima_aceh.fittedvalues
```

```python
pred_aceh = model_arima_aceh.forecast(steps=2)[0]
pred_aceh
```

```python
mse = mean_squared_error(data_aceh[1:], model_arima_aceh.fittedvalues)
mse
```

```python
data_aceh.plot(figsize=(15,12))
plt.plot(model_arima_aceh.fittedvalues,color='red')
```

```python
model_arima_aceh.plot_predict(3,25)
```
